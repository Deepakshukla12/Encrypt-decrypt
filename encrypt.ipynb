```python
from google.colab import files
from docx import Document
from cryptography.fernet import Fernet
import io

# Function to generate a key
def generate_key():
    return Fernet.generate_key()

# Function to encrypt a message
def encrypt_message(message, key):
    cipher_suite = Fernet(key)
    encrypted_message = cipher_suite.encrypt(message.encode())
    return encrypted_message

# Function to decrypt a message
def decrypt_message(encrypted_message, key):
    cipher_suite = Fernet(key)
    decrypted_message = cipher_suite.decrypt(encrypted_message).decode()
    return decrypted_message

# Prompt the user to upload a file
uploaded = files.upload()

# Get the first uploaded file name
file_name_to_access = next(iter(uploaded))

# Access the content of the uploaded file
if file_name_to_access:
    # Generating key
    key = generate_key()

    document = Document(io.BytesIO(uploaded[file_name_to_access]))

    # Read the content of the Word document
    file_content = ""
    for paragraph in document.paragraphs:
        file_content += paragraph.text + '\n'

    # Encrypt the content
    encrypted_content = encrypt_message(file_content, key)
    print("\n")

    print(f'Encrypted content:\n{encrypted_content}')

    # Decrypt the content for demonstration purposes
    decrypted_content = decrypt_message(encrypted_content, key)
    print(f'Decrypted content:\n{decrypted_content}')

    print("\nMessage Encryption and Decryption is done")


else:
    print('Error: No files uploaded.')
```
